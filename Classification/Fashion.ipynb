```python
import numpy as np
import pandas as pd
import os
!pip install -U efficientnet
dir_path="/content/drive/MyDrive/fabric_madtar"
classname="tar"
for dirname, _, filenames in os.walk(dir_path):
    for filename in filenames:
        print(os.path.join(dirname, filename))
from keras.models import Model
from tensorflow.keras.applications import EfficientNetV2S
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.efficientnet import preprocess_input
base_model = EfficientNetV2S(weights='imagenet')
model = Model(inputs=[base_model.input],outputs=[base_model.get_layer('predictions').output])
model.summary()
```

```python
import tensorflow as tf
from glob import glob
from sklearn.preprocessing import MinMaxScaler
parent_directory = '/content/drive/MyDrive/fabric_madtar/'
imagePatches = glob(dir_path+'/'+classname+'/*.png', recursive=True)
imagename=[]
image = []
for subfolder in glob(parent_directory + '*/'):
    subfolder_name = subfolder.split('/')[-2]  # Extract the subfolder name
    imagePatches = glob(subfolder + '*.JPG')
    for img in imagePatches:
            imagename.append(img)
            img_data = tf.keras.utils.load_img(img, target_size=(384, 384))
            img_data = tf.keras.utils.img_to_array(img_data)
            features = []
            img_data = np.expand_dims(img_data, axis=0)
            img_data = preprocess_input(img_data)
            feats = model.predict(img_data)
            features.append(feats.flatten())
            data = feats.flatten()
            image.append(data)
    # scaler = MinMaxScaler()
    # print(image)
    # scaler.fit(image)
    # features = scaler.transform(image)
    data = pd.DataFrame(image)
    csv_path=dir_path+'/'+classname+'.csv'
    data.insert(0, 'Image_name', imagename)
    print(imagename)
    print(data)
    data.to_csv("/content/"+subfolder_name+".csv",index=False)
```

```python
model = YOLO('/content/drive/MyDrive/runs/detect/train/weights/best.pt')
results = model('/content/new/download.jpeg')
for r in results:
  boxes = r.boxes
  names = r.names
tensor_value = str(int(boxes.cls.item()))
print("The class is:",names[int(tensor_value)])
k=names[int(tensor_value)]
```

```python
csv_path_det = '/content/single.csv'
data_det = pd.read_csv(csv_path_det)
image_names_det = data_det['Image_name']
image_features_det = data_det.drop(['Image_name'], axis=1)
a1=image_features_det.values.tolist()
```

```python
target_path = "/content/"+k+".csv"
data_tar = pd.read_csv(target_path)
image_names_tar = data_tar['Image_name']
img_n = image_names_tar.values.tolist()
image_features_tar = data_tar.drop(['Image_name'], axis=1)
a2=image_features_tar.values.tolist()
for i in a2:
  i.pop(-1)
euc_dist=[]
for i in a2:
  distance=0
  img=[]
  for j in range(len(i)):
    distance += (a1[0][j] - i[j]) ** 2
  img.append(distance)
  euc_dist.append(img)
for i in range(len(img_n)):
  euc_dist[i].append(img_n[i])
```
